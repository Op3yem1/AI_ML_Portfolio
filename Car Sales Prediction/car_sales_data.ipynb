## Messy Car Sales Data

Using Python, I create a csv file with deliberately incomplete or incorrect data here to use further in this notebook.

```python
import csv
import random
import pandas as pd

# Define possible values
makes = ["toyota", "Ford", "BMW", "Honda", "Mercedes", "Audi", "Volkswagen", "Hyundai", "Nissan", "Peugeot","Seat","Citreon",None]
colours = ["Red", "Blue", "Green", "Black", "White", "Silver", "Grey", "Yellow", "Aqua", "Purple", None]
doors_options = [3.0, 4.0, 5.0, 7.0, None]
k = 800000000      # Arbitrary constant: 8 x 10^8

# Generate dummy data
data = []
for _ in range(500):
    make = random.choice(makes)
    colour = random.choice(colours)
    doors = random.choice(doors_options)
    mileage = round(random.uniform(10000.0, 160000.0))
    if mileage > 65000:
      margin = random.randint(-1000, 1000)  # The margin is added or subtracted from the calc_price to allow for fluctuation in prices
    else:
      margin = random.randint(-3000, 3000)
    calc_price = round((k/mileage) + margin)  # The calculated price is inversely proportional to the mileage using an arbitrary constant
    price = "{:,.2f}".format(calc_price)
    data.append([make, colour, mileage, doors, price])

# Write to CSV file
filename = "car-sales-messy-data.csv"
with open(filename, mode="w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["Make", "Colour", "Milage", "Doors", "Price"])
    writer.writerows(data)

print(f"CSV file '{filename}' created successfully!")
```

## Training a Regression Model

First, I split the data into training and test sets and then train it using sci-kit learn

```python
# import module from sklearn
from sklearn.model_selection import train_test_split

x = tidy_car_sales[['Mileage']]
y = tidy_car_sales['Price ()']

# 80:20 traning:test split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42) # random state produces pseudo random data that can be reproduced
```

```python
x.shape          # check that x is 2D array
```

```python
x_train.head(), y_train.head()
```

```python
# Train the model
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train, y_train)

print(f'Intercept: {model.intercept_}')
print(f'Slope: {model.coef_[0]}')
```

## Evaluating and Visualising the Results

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Predict prices
y_pred = model.predict(x_test)

# Print evaluation metrics
print("Mean Absolute Error:", mean_absolute_error(y_test, y_pred))
print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
print("R2 Score:", r2_score(y_test, y_pred))
```

MAE indicates that the average prediction price is off by about 6,100. Since this is quite a high number, it suggests that mileage alone is not the only factor affecting the prices.
R2 score shows that 54% of the variation in car prices is due to mileage. A good target R2 score would be about 0.75

```python
import matplotlib.pyplot as plt

plt.scatter(x_test, y_test, color="blue", label="Actual Data")
plt.plot(x_test, y_pred, color="red", label="Regression Line")
plt.xlabel("Mileage")
plt.ylabel("Price ()")
plt.legend()
plt.show()
```
