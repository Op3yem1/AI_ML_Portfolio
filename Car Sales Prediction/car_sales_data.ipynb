## Messy Car Sales Data

Using Python, I create a csv file with deliberately incomplete or incorrect data here to use further in this notebook.

```python
import csv
import random

# Define possible values
makes = ["toyota", "Ford", "BMW", "Honda", "Mercedes", "Audi", "Volkswagen", "Hyundai", "Nissan", "Peugeot","Seat","Citreon",None]
colours = ["Red", "Blue", "Green", "Black", "White", "Silver", "Grey", "Yellow", "Aqua", "Purple", None]
doors_options = [3.0, 4.0, 5.0, 7.0, None]
k = 800000000      # Arbitrary constant: 8 x 10^8

# Generate dummy data
data = []
for _ in range(500):
    make = random.choice(makes)
    colour = random.choice(colours)
    doors = random.choice(doors_options)
    mileage = round(random.uniform(10000.0, 160000.0))
    if mileage > 65000:
      margin = random.randint(-1000, 1000)  # The margin is added or subtracted from the calc_price to allow for fluctuation in prices
    else:
      margin = random.randint(-3000, 3000)
    calc_price = round((k/mileage) + margin)  # The calculated price is inversely proportional to the mileage using an arbitrary constant
    price = "{:,.2f}".format(calc_price)
    data.append([make, colour, mileage, doors, price])

# Write to CSV file
filename = "car-sales-messy-data.csv"
with open(filename, mode="w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["Make", "Colour", "Milage", "Doors", "Price"])
    writer.writerows(data)

print(f"CSV file '{filename}' created successfully!")
```
